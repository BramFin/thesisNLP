```python
#remove newline 
df_reduced['item_7'] = df_reduced['item_7'].replace(r'\r+|\n+|\t+',' ', regex=True)
#lowcase the words
df_reduced['item_7']=df_reduced['item_7'].str.lower()

#check if the extraction went correct if item 7. not at index 0 something went wrong
sub="item 7."
df_reduced["Indexes"]= df_reduced["item_7"].str.find(sub)

#remove extra white spaces
df_new['item_7'] = df_new['item_7'].str.replace('  ', ' ') # Remove extra spaces

#remove companies were extraction went wrong
df_new = df_reduced[df_reduced['Indexes'] == 0]
```

```python
# tokenize the text for loughran mcdonald text processing
import nltk
from nltk.tokenize import sent_tokenize
df_new['tokenized_text']=df_new['item_7'].apply(sent_tokenize)

#explode text so every sentence is a row
df_new=df_new.explode('tokenized_text')
```

```python
#loughran and mcdonald how many positive and negative words in sentence

lm=ps.LM()
df_new1['tokenized_text'] = df_new1['tokenized_text'].astype(str)
df_new1['tokenize'] = df_new1['tokenized_text'].apply(lambda x: lm.tokenize(x))
df_new1['positive_lm_MDA']=df_new1['tokenize'].apply(lambda x: lm.get_score(x)['Positive'])
df_new1['negative_lm_MDA']=df_new1['tokenize'].apply(lambda x: lm.get_score(x)['Negative'])
```

```python
#classify the sentence when more positive words positive sentence vice versa negative and equal is neutral

#Positive
df_new1['classification_lm'] = 'positive'
#negative
df_new1.loc[df_new1['positive_lm_MDA'] < df_new1['negative_lm_MDA'], 'classification_lm'] = 'negative'
#neutral
df_new1.loc[df_new1['positive_lm_MDA'] == df_new1['negative_lm_MDA'], 'classification_lm'] = 'neutral'
```

```python
#create new table with the tone and tone normalized for every company
data=df_new1.groupby('cik')['classification_lm'].value_counts().reset_index(name='counts')
#pivot data
data=data.pivot_table(index="cik", columns="classification_lm", values="counts", aggfunc=np.sum)
#calculate the tone
data['tone_lm']=(data['positive']-data['negative'])/(data['positive']+data['negative'])
#normalize the tone
data['tone_lm_normalized'] = preprocessing.scale(data['tone_lm'])
data
```
