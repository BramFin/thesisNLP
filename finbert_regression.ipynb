```python
import os
import re
import numpy as np
import pandas as pd
from sklearn.preprocessing import scale

# Plotting
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# cik to ticker number dataset

cik=pd.read_csv('cik.txt', sep='\t',header=None,)
cik=cik.rename(columns={0:'ticker',1:'cik'})
cik['ticker']=cik['ticker'].str.upper()
#industry dataset
industry = pd.read_csv('gvacwzxxaku648ow.csv')
#revenue dataset
revenue= pd.read_csv('revenue.csv')
revenue['year'] = pd.DatetimeIndex(revenue['date']).year
# ratios data roe boek to market
ratios = pd.read_csv('ratios.csv')
ratios['month_year'] = pd.to_datetime(ratios['date']).dt.to_period('M')
#insert the beta dataset
betas= pd.read_csv('0hbtg2fwhrp54kma.csv')

betas=betas.rename(columns={'DATE':'filing_date','TICKER':'ticker'})
betas['filing_date']=pd.to_datetime(betas['filing_date'])
betas


stock_returns= pd.read_csv('mpfpz39cfum72pwp.csv')
stock_returns=stock_returns.rename(columns={'TICKER':'ticker'})
stock_returns['date']=pd.to_datetime(stock_returns['date'])

```

```python

df_fin=pd.read_csv('finbert_2021.csv')
df_fin
```

```python
vix = pd.read_csv('vix.csv')
df_fin['filing_date']=pd.to_datetime(df_fin['filing_date'])
vix['filing_date']=pd.to_datetime(vix['filing_date'])
df_fin=df_fin.merge(vix, on='filing_date', how='inner')
df_fin.dropna()
```

```python

df_fin['normalized_all']=scale(df_fin['tone_finbert'])
df_fin['length']=df_fin['negative']+df_fin['positive']+df_fin['neutral']
```

```python
df_fin['year'] = pd.DatetimeIndex(df_fin['filing_date']).year
df_fin['filing_date'] = pd.to_datetime(df_fin['filing_date'])
df_fin['quarter'] = pd.PeriodIndex(df_fin['filing_date'], freq='Q')
df_fin=df_fin.dropna()
df_fin
```

```python
data3=df_fin.merge(cik, on='cik', how='inner')
data3.drop_duplicates(subset=['cik','year'], keep='first', inplace=True, ignore_index=True)
```

```python
#create dates for the event study
data3['end']= data3['filing_date']+pd.tseries.offsets.BusinessDay(-1)# end date 
data3['start']= data3['filing_date']+pd.tseries.offsets.BusinessDay(-1)#start date -1 is event date minus one
```

```python
data3['date'] = list(map(lambda x, y: pd.date_range(start=x, end=y),
                      data3['start'], 
                      data3['end']))
data3 = (data3.explode('date'))
```

```python
#merge the data3  with the stock returns
data3['date']=pd.to_datetime(data3['date'])
merged3=data3.merge(stock_returns, on=['ticker','date'], how='inner')

```

```python
merged3['AR']=merged3['RETX'].astype(float)-merged3['sprtrn']
```

```python
merged4=merged3.merge(betas, on=['filing_date','ticker'], how='left')
```

```python
merged4['CAR']=merged4['AR']*merged4['b_mkt']
```

```python
abnormalreturns=merged4.groupby(['filing_date','vix','ticker','tone_finbert','normalized_all','tone_finbert_normalized','length'])['CAR','AR'].sum()
abnormalreturns=pd.DataFrame(abnormalreturns).reset_index()
abnormalreturns['quarter'] = pd.PeriodIndex(abnormalreturns['filing_date'], freq='Q')
```

```python
#add ratios
#ratios['year']=pd.DatetimeIndex(ratios['date']).year
ratios['month_year'] = pd.to_datetime(ratios['date']).dt.to_period('M')

abnormalreturns['month_year'] = pd.to_datetime(abnormalreturns['filing_date']).dt.to_period('M')
abnormalreturns= abnormalreturns.merge(ratios, on=['ticker','month_year'], how='inner')

```

```python
revenue['year'] = pd.DatetimeIndex(revenue['date']).year

abnormalreturns['year']=pd.DatetimeIndex(abnormalreturns['filing_date']).year
abnormalreturns= abnormalreturns.merge(revenue, on=['ticker','year'], how='inner')
```

```python
y = abnormalreturns['CAR']
X = abnormalreturns[[]]
X = sm.add_constant(X)
model11 = sm.OLS(y, X.astype(float)).fit()
```
